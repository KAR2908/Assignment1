Creating random vector of size 15 with integers in range 1-20

```python
v1 = np.random.randint(1, 21, size=15)
print(v1)
```

Reshaping the array size to 3 by 5

```python
v2 = v1.reshape(3,5)
print(v2)
```

Replace the max in each row by 0

```python
max_indices = np.argmax(v2, axis=1)
v2[np.arange(len(v2)),max_indices] = 0
print(v2)
```

Program to compute the eigenvalues and eigenvectors of a given square array given below:[[3 -2][1 0]]

```python
#Array declaration
array1 =  np.array([[3, -2], [1, 0]])

#compute eigen values and eigenvectors
eigenval, eigenvect = np.linalg.eig(array1)

#printing the results:
print("Eigenvalues are:", eigenval)
print("Right eigenvectors are: ")
for i, eigenvect in enumerate(eigenvect.T):
    print(f"Eigenvector {i+1}:", eigenvect)
```

Sum of the diagonal elements of a given array:[[0 1 2][3 4 5]]

```python
#Array declaration
arr2 = np.array([[0, 1, 2], [3, 4, 5]])
#sum of diagonal elements
dg_sum = np.trace(arr2)
#print the sum
print("sum of diagonal elements is", dg_sum)
```

program to create new shape to array without changing its data. Reshape 3x2 :[[1 2][3 4][5 6]]

```python
#array declaration
array3= np.array([[1, 2], [3, 4], [5, 6]])
reshaped_array3 = array3.reshape((2, 3))
print("Reshaped 3x2 array: ")
print (reshaped_array3)
```

Reshape 2X3: [[1 2 3][4 5 6]]

```python
array4= np.array([[1, 2, 3], [4, 5, 6]])
reshaped_array5 = array4.reshape((3, 2))
print("Reshaped 2x3 array: ")
print (reshaped_array5)
```

2. Sample data:
Programming languages: Java, Python, PHP, JavaScript, C#, C++
